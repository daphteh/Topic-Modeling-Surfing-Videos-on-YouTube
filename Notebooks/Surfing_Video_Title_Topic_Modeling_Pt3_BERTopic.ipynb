```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('/content/video_df.csv')
docs = dataset['title']
```

```python
from bertopic import BERTopic

topic_model1 = BERTopic(language="english", calculate_probabilities=True, verbose=True)
topics1, probs1 = topic_model1.fit_transform(docs)
```

```python
from bertopic.representation import KeyBERTInspired

# Fine-tune your topic representations
representation_model = KeyBERTInspired()
topic_model2 = BERTopic(representation_model=representation_model)
topics2, probs2 = topic_model2.fit_transform(docs)
```

```python
freq1 = topic_model1.get_topic_info()
freq2 = topic_model2.get_topic_info()
```
